# Using Transencoder

```python
df=df.groupby('Transaction')['Item'].apply(list)
```

```python
df.head()
```

```python
df1=df
```

```python
type(df1)
```

```python
df1=df1.values.tolist()
```

```python
df1
```

```python
from mlxtend.preprocessing import TransactionEncoder
te=TransactionEncoder()
```

```python
arr=te.fit(df1).transform(df1)
df=pd.DataFrame(arr,columns=te.columns_)
```

```python
df.head()
```

```python
frequent_itemsets = apriori(df, min_support=0.01 ,use_colnames=True)
rules = association_rules(frequent_itemsets ,metric="lift")
rules.sort_values('lift' ,ascending=False, inplace=True)
rules.head(10)
```

```python
frequent_itemsets.sort_values('support' ,ascending=False, inplace=True)
frequent_itemsets.head()
```

```python
rules['Length']=frequent_itemsets['itemsets'].apply(lambda X:len(X))
rules
```

```python
rules[(rules['Length'] == 2) & (rules['lift'] >= 1)]
```

```python
l=frequent_itemsets[frequent_itemsets['itemsets'] == {'Coffee','Cake'}]
m=frequent_itemsets[frequent_itemsets['itemsets'] == {'Coffee','Tea','Cake'}]
l
```

# Working with sparse matrix

```python
from mlxtend.preprocessing import TransactionEncoder
te=TransactionEncoder()
arr=te.fit(df1).transform(df1,sparse=True)
df=pd.DataFrame.sparse.from_spmatrix(arr,columns=te.columns_)
```

```python
df
```

```python
frequent_itemsets = apriori(df, min_support=0.01 ,use_colnames=True)
rules = association_rules(frequent_itemsets ,metric="lift")
rules.sort_values('confidence' ,ascending=False, inplace=True)
rules
```
